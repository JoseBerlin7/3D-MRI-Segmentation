# Extracting ZIP Files

```python
import os
import zipfile
```

```python
location = "./data"
for file in os.listdir(location):
    if file.endswith(".zip"):
        file_path = os.path.join(location, file)
        with zipfile.ZipFile(file_path, "r") as zip_ref:
            zip_ref.extractall(location)
        print(f"Extracted: {file_path}\n Removing zip file... ", end=" ")
        os.remove(file_path)
        print(" Done.")
```

# Files Info

```python

for dir in os.listdir(location):
    dir_path = os.path.join(location, dir)
    print(len(os.listdir(dir_path)) if os.path.isdir(dir_path) else "Not a", "file in ", dir)
```
